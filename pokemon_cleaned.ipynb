```python
# Dependencies
import pandas as pd
```

```python
file = 'pokemon (1).csv'
```

```python
poke_csv = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
#create DataFrame from csv
poke_df = pd.DataFrame(poke_csv)
poke_df
```

```python
#drop unneeded columns and columns with missing data
poke_df.drop(['percentage_male', 'is_legendary', 'base_egg_steps', 'base_happiness', 'capture_rate', 'japanese_name', 'classfication', 'experience_growth', 'height_m', 'weight_kg'], axis=1, inplace=True)
poke_df
```

```python
#clean up abilities column
poke_df['abilities'] = poke_df['abilities'].str.strip('[]')
ability_df = poke_df['abilities'].str.split(', ', n=-1, expand=True)

#split and join to main DataFrame
ability_df = ability_df[[0,1]]
ability_df.rename(columns={"0": "ability1", "1": "ability2"})
poke_df = poke_df.join(ability_df)
```

```python
#rename and clean new ability columns
poke_df = poke_df.rename(columns={0: 'ability1', 1: 'ability2'})
poke_df['ability1'] = poke_df['ability1'].str.strip("''")
poke_df['ability2'] = poke_df['ability2'].str.strip("''")
poke_df
```

```python
#drop original abilities column
poke_df.drop(['abilities'], axis=1, inplace=True)
poke_df.columns
```

```python
#reorder columns
poke_df = poke_df[['pokedex_number','name','generation','type1','type2','ability1','ability2','base_total','hp','attack','defense','sp_attack','sp_defense','speed','against_normal','against_fire','against_water','against_grass','against_electric','against_rock','against_ground','against_flying', 'against_fight','against_ice','against_bug','against_steel','against_poison','against_psychic','against_dark','against_ghost','against_fairy','against_dragon']]

#tried to create a loop to handle this at first, but no luck
poke_df = poke_df.rename(columns = {'against_normal':'normal','against_fire':'fire','against_water':'water','against_grass':'grass','against_electric':'electric','against_rock':'rock','against_ground':'ground','against_flying':'flying','against_fight':'fighting','against_ice':'ice','against_bug':'bug','against_steel':'steel','against_poison':'poison','against_psychic':'psychic','against_dark':'dark','against_ghost':'ghost','against_fairy':'fairy','against_dragon':'dragon'})

#set index as Pokedex number
poke_df = poke_df.set_index('pokedex_number')
poke_df
```

```python
#replace NaN and blanks with "-" in the type2 and ability2 columns
poke_df = poke_df.fillna(value='-')
poke_df
```

```python
poke_df.to_csv("pokedata_cleaned.csv", index=True, header=True)
```
